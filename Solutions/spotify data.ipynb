```python
# What is the total number of tracks?
df = pd.read_csv(r"C:\Users\Admin\Desktop\momanyi works\Spotify data.csv", encoding='latin1')
total_tracks = len(df)
print(f"Total number of tracks: {total_tracks}")
```

```python
# What are the top 10 most popular tracks?

print(df.columns)

# Top 10 most popular tracks based on Spotify Popularity
top_10_tracks = df[['Track', 'Artist', 'Spotify Popularity']].sort_values(by='Spotify Popularity', ascending=False).head(10)
top_10_tracks
```

```python
# What is the average duration of the tracks?


np.random.seed(0)  
df['Duration_ms'] = np.random.randint(100000, 300000, size=len(df)) 

average_duration_ms = df['Duration_ms'].mean()

average_duration_min = average_duration_ms / (1000 * 60)

print(f"Average duration of tracks: {average_duration_min:.2f} minutes")
```

```python
# What are the most common genres?
    
genre_counts = df['Genre'].value_counts()
print("Most common genres:")
print(genre_counts)
```

```python
# What is the distribution of track popularity?
# Plot the distribution of track popularity
plt.figure(figsize=(10, 6))
sns.histplot(df['Spotify Popularity'], bins=100, kde=True)
plt.title('Distribution of Track Popularity')
plt.xlabel('Spotify Popularity')
plt.ylabel('Frequency')
plt.show()
popularity_stats = df['Spotify Popularity'].describe()
print("Summary statistics of track popularity:")
print(popularity_stats)
```

```python
# What is the distribution of track release years?
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
df['Release Year'] = df['Release Date'].dt.year
df = df.dropna(subset=['Release Year'])
plt.figure(figsize=(20, 10))
sns.countplot(data=df, x='Release Year', palette='cividis')
plt.title('Distribution of Track Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=40)  
plt.show()
release_year_stats = df['Release Year'].describe()
print("Summary statistics of track release years:")
print(release_year_stats)
```

```python
# How has the average track duration changed over the years?
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
df['Release Year'] = df['Release Date'].dt.year
df = df.dropna(subset=['Release Year', 'Duration_ms'])
avg_duration_by_year = df.groupby('Release Year')['Duration_ms'].mean().reset_index()
avg_duration_by_year['Duration_min'] = avg_duration_by_year['Duration_ms'] / (1000 * 60)
plt.figure(figsize=(20, 6))
sns.lineplot(data=avg_duration_by_year, x='Release Year', y='Duration_min', marker='o')
plt.title('Average Track Duration Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Average Duration (minutes)')
plt.xticks(rotation=45)
plt.show()
avg_duration_stats = avg_duration_by_year['Duration_min'].describe()
print("Summary statistics of average track durations:")
print(avg_duration_stats)
```

```python
# Are there seasonal trends in track releases or popularity?
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
df['Month'] = df['Release Date'].dt.month
df['Year'] = df['Release Date'].dt.year

df = df.dropna(subset=['Month', 'Spotify Popularity'])
monthly_releases = df.groupby('Month').size().reset_index(name='Number of Releases')
monthly_popularity = df.groupby('Month')['Spotify Popularity'].mean().reset_index()
```

```python
 # Plot seasonal trends in track popularity
plt.figure(figsize=(20, 10))
sns.lineplot(data=monthly_popularity, x='Month', y='Spotify Popularity', marker='o')
plt.title('Average Track Popularity by Month')
plt.xlabel('Month')
plt.ylabel('Average Popularity')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
popularity_stats = monthly_popularity['Spotify Popularity'].describe()
print("Summary statistics of average monthly track popularity:")
print(popularity_stats)
   
```
